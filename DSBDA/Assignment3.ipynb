```python
def get_stats(data: pd.DataFrame | pd.Series, column_name: str = ""):
    if type(data) == pd.DataFrame:
        describe_dict = {}
        for col in data.columns:
            if (data[col].dtype == np.int64) or (data[col].dtype == np.float64):
                column = data[col]
                count = column.count()
                mean = column.mean()
                std = column.std()
                min = column.min()
                quant_25 = column.quantile(0.25)
                quant_50 = column.quantile(0.5)
                quant_75 = column.quantile(0.75)
                max = column.max()
                describe_dict[str(col)] = np.array([count, mean, std, min, quant_25, quant_50, quant_75, max])
    elif type(data) == pd.Series:
        describe_dict = {}
        if (data.dtype == np.int64) or (data.dtype == np.float64):
            column = data
            count = column.count()
            mean = column.mean()
            std = column.std()
            min = column.min()
            quant_25 = column.quantile(0.25)
            quant_50 = column.quantile(0.5)
            quant_75 = column.quantile(0.75)
            max = column.max()
            describe_dict[data.name] = np.array([count, mean, std, min, quant_25, quant_50, quant_75, max])
    if column_name != "":
        return pd.DataFrame(describe_dict, index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])[column_name]
    return pd.DataFrame(describe_dict, index=['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'])
```

```python
describe_df = get_stats(income_data)
describe_df
```

```python
def ask(data: pd.DataFrame | pd.Series):
    if type(data) == pd.DataFrame:
        while True:
            choice = input("Enter Column name : ")
            if choice in data.columns:
                print(get_stats(data=data[choice]))
            elif choice == "":
                print(get_stats(data=data))
            elif (choice == "X") or (choice == "x"):
                print("Bye")
                break
            elif choice == "--columns":
                print(list(data.columns))
            elif (choice == "--help") or (choice == "-h") or (choice == "help"):
                print("Enter '--columns' to see columns.\nEnter 'X' to quit.")
            else:
                print(f"No such option. \nEnter --help or -h to get list of commands.")
                if data.columns.str.contains(choice.upper()).any():
                    print(f"\nDid you mean {choice.upper()}?")
                elif data.columns.str.contains(choice.capitalize()).any():
                    print(f"\nDid you mean {choice.capitalize()}?")
    elif type(data) == pd.Series:
        print(get_stats(data))
    else:
        raise Exception(f"Invalid data type {type(data)}. Please pass DataFrame or Series object.")
```
